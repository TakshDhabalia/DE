```python
#Demo of Feature Encoding also converting categorical data into numbers
import pandas as pd
info = {
    'Gender' : ['male', 'female', 'female', 'male', 'female', 'female'],
    'Position' : ['CEO', 'Cleaner', 'Employee', 'Cleaner', 'CEO', 'Cleaner']
}
df = pd.DataFrame(info)
print(df)
```

```python
#A LabelEncoder converts a categorical data into a number ranging from 0 to n-1,
#where n is the number of classes in the variable.
#These are represented as 0,1,2 in alphabetical order.
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
gender_encoded = le.fit_transform(df['Gender'])
print(gender_encoded)
```

```python
df['encoded_gender'] = gender_encoded
print(df)
```

```python
encoded_position = le.fit_transform(df['Position'])
df['encoded_position'] = encoded_position
print(df)
```

```python
normalized_price = (mobile_wt - mobile_wt.min())/(mobile_wt.max()-mobile_wt.min())
data['normalized_weight']=normalized_price
data.sort_values(by='mobile_wt')
```

A Min-Max scaling is typically done via the following equation:

Xsc=X−XminXmax−Xmin.

```python
from sklearn import preprocessing
```

```python
x = data[['ram']].values.astype(float)
x.max()
```

```python
# Create a minimum and maximum processor object
min_max_scaler = preprocessing.MinMaxScaler()
# Create an object to transform the data to fit minmax processor
x_scaled = min_max_scaler.fit_transform(data[['ram']].values.astype(float))
# Run the normalizer on the dataframe
df_normalized = pd.DataFrame(x_scaled)
data['normalized2']=x_scaled
data.sort_values(by='ram')
```

**Z-Score Normalization**

```python
# Convert 'ram' column to numeric type before calculating z-score
data['ram'] = pd.to_numeric(data['ram'], errors='coerce')

# Calculate z-score
from scipy import stats
data['zscore'] = stats.zscore(data['ram'])

data.sort_values(by='ram')
```

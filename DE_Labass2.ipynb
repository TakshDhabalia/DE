```python
# loading csv file
sample = pd.read_csv('index.csv')
print(sample)
```

```python
# info
print(sample.info())
```

```python
# describe
print(sample.describe())
```

```python
# identifying missing values
print(sample.isnull())
```

```python
# returning total number of null values
print(sample.isnull().sum())
```

```python
# Drop all rows that have any NaN (missing) values
sample.dropna()
```

```python
# Drop only if the entire row has NaN (missing) values.
sample.dropna(how ='all')
```

```python
# Drop only if a row has more than two NaN (missing) values.
sample.dropna(thresh=sample.shape[1] - 2)
```

```python
# Replace NaN values with the default value (e.g., '0')
sample_filled = sample.fillna('0')
sample_filled
```

```python
# Calculating the mean for a specific column (e.g., 'column_name')
mean_value = sample['money'].mean()
sample['money'] = sample['money'].fillna(mean_value)
```

```python
# Identify duplicate rows
duplicates = sample.duplicated()
sample[duplicates]
```

```python
duplicates.sum()
```

```python
# removing duplicates
sample_no_duplicates = sample.drop_duplicates()
sample_no_duplicates
```
